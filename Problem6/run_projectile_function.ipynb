```python
import motion
```

#### Typical Input

```python
motion.projectile(30, 30, "m/s", 20)
```

#### Error: Angle <0 or >90

```python
motion.projectile(30, 120, "m/s", 20)
```

#### Error: Velocity is negative

```python
motion.projectile(-30, 30, "m/s", 20)
```

#### Error: Initial height is negative

```python
motion.projectile(30, 30, "m/s", -20)
```

#### Error: Units and initial height are switched
The inputs are switched back to their correct variables so the program still runs correctly

```python
motion.projectile(30, 30, 20, "m/s")
```

#### Error: Launch angle and units are switched
The inputs are switched back to their correct variables so the program still runs correctly

```python
motion.projectile(30, "m/s", 30, 20)
```
